```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

path = 'https://raw.githubusercontent.com/diplomado-bigdata-machinelearning-udea/Curso1/master/s04/'
dfE = pd.read_csv(path+'ESTADO_201709.txt', delimiter="|")
dfT = pd.read_csv(path+'TRANSACCION_201709.txt', delimiter="|")

print(dfE.head())
print(dfT.head())
```

```python
dfE.columns = dfE.columns.str.strip()
dfT.columns = dfT.columns.str.strip()
```

```python
df_merge = pd.merge(dfE, dfT, on='NUMERO TARJETA', how='inner')
```

```python
print(df_merge.shape)  # Número de filas y columnas
print(df_merge.head())
```

```python
# Quitar espacios de los nombres de las columnas
dfE.columns = dfE.columns.str.strip()
dfT.columns = dfT.columns.str.strip()

# Convertir automáticamente todas las columnas que contengan "FECHA"
for col in dfE.columns:
    if "FECHA" in col.upper():
        dfE[col] = pd.to_datetime(dfE[col], errors='coerce')

for col in dfT.columns:
    if "FECHA" in col.upper():
        dfT[col] = pd.to_datetime(dfT[col], errors='coerce')
```
