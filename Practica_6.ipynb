```python
# Inicializamos la variable
resultado = 0.0

# Pedimos el primer número
primer_numero = float(input("Digite el primer número: "))

# Pedimos el operador
operador = input("Digite el operador (+, -, *, /): ")

# Validamos el operador
if operador not in ["+", "-", "*", "/"]:
    print("Operador no válido. Por favor, elija +, -, * o /.")
else:
    # Pedimos el segundo número
    segundo_numero = float(input("Digite el segundo número: "))

    # Calculamos el resultado según el operador
    if operador == "+":
        resultado = primer_numero + segundo_numero
    elif operador == "-":
        resultado = primer_numero - segundo_numero
    elif operador == "*":
        resultado = primer_numero * segundo_numero
    elif operador == "/":
        if segundo_numero == 0:
            print("❌ Error: No se puede dividir por cero.")
        else:
            resultado = primer_numero / segundo_numero

    # Mostrar el resultado (solo si no hubo error)
    if operador != "/" or segundo_numero != 0:
        print("✅ El resultado es:", resultado)
```

```python
#Juego Adivinar el numero
import random
numero_secreto = random.randint(1, 100)
intentos = 0
print("¡Bienvenido al juego de adivinar el número!")
while True:
    intentos += 1
    try:
        numero_usuario = int(input("Adivina el número (entre 1 y 100): "))
    except ValueError:
        print("Por favor, ingresa un número válido.")
        continue
    if numero_usuario < numero_secreto:
        print("El numero es muy bajo intenta con uno mas alto")
    elif numero_usuario > numero_secreto:
        print("El numero es muy alto intenta con uno mas bajo")
    else:
        print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
        break
```
